## Missing values in Categorical Features

```python
categorical_null=[i for i in categorical_variables if dataset[i].isnull().sum()>1]
```

```python
len(categorical_null)
```

### Filling with some new label 'Missing'

```python
def replace_with_missing(dataset,categorical_null):
    data=dataset.copy()
    data[categorical_null]=data[categorical_null].fillna('Missing')
    return data
dataset=replace_with_missing(dataset,categorical_null)
```

```python
dataset[categorical_null].isnull().sum()
```

## Missing values of Numerical_variables

```python
numerical_null=[i for i in numerical_variables if dataset[i].isnull().sum()>1]
```

```python
dataset[numerical_null].isnull().sum()
```

### Handling the Numerical Missing variables`

```python
for i in numerical_null:
    average=dataset[i].median()
    dataset[i+'nan']=np.where(dataset[i].isnull(),1,0)
    dataset[i].fillna(average,inplace=True)
```

```python
dataset[numerical_null].isnull().sum()
```

## Modifying year_feature 
they are temporal variables

```python
for i in year_feature:
    if i!='YrSold':
        dataset[i]=dataset['YrSold']-dataset[i]
```

## Log transformation
We having the skewed features in continuous features so we have to apply log transform.
#### (the continuous_featues which do not contains 0 in their unique records can be considered as Skewed features) 

```python
logtrans_variables=[i for i in continuous_features if 0 not in dataset[i].unique()]
```

```python
logtrans_variables
```

```python
dataset[logtrans_variables].head()
```

```python
for i in logtrans_variables:
    dataset[i]=np.log(dataset[i])
```

```python
dataset[logtrans_variables].head()
```

## Handling the Rare categorical features Unique records < 1% or 0.01 

```python
for i in categorical_variables:
    temp=dataset.groupby(i)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[i]=np.where(dataset[i].isin(temp_df),dataset[i],'Rare_var')
```

```python
dataset[categorical_variables].head()
```

## Converting the Object/Categorical features into Numerical

```python
categorical_features=[i for i in dataset.columns if dataset[i].dtype=='O']
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head(10)
```

# Feature Scaling

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],axis=1)
```
